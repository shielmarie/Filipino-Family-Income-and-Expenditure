```python
#Detection of outliers. 
#I do not have to remove the outliers because the data is correct and is needed to create meaningful results.

q1_income, q3_income= np.percentile(df['Total Household Income'],[25,75])
IQR= q3_income- q1_income
print ("IQR\n", IQR)

lb= q1_income- 1.5*IQR
ub= q3_income+ 1.5*IQR

print ("Lowerbound\n", lb)
print ("Upprerbound\n", ub)
```

```python
#Finding 1: Highest outlier values are coming from Region VI
sns.boxplot (y="Region", x="Total Household Income", data=df)
```

```python
#Finding 2: Region VI has the highest total household income in the Philippines. Main source of income is other sources of income.
#Finding 3: Interestingly enough, the main source of income for the lowest total household income 
#are mostly coming from other sources of income as well.
#The highest and lowest income are both coming from other sources of income.
sns.scatterplot(y="Region", x="Total Household Income", data=df,hue="Main Source of Income")
```

```python
#Finding 4: Although Region VI has the highest total household income, Region VII has the highest total food expenditure.
sns.scatterplot(y="Region", x="Total Food Expenditure", data=df, hue="Main Source of Income")
```

```python
#Finding 5: The top 3 highest spenders for rice are Region VII, Region VI, and Region XII. 
#The highest spenders for rice are coming from all 3 main source of income.
sns.scatterplot(y="Region", x="Total Rice Expenditure", data=df, hue="Main Source of Income")
```

```python
#Finding 6: The top 3 highest spenders for meat are Caraga, Region VIII, and Region II. 
#The highest spenders for meat are coming from all 3 main source of income. 
sns.scatterplot(y="Region", x="Meat Expenditure", data=df, hue="Main Source of Income")
```

```python
#Finding 7: The top 3 highest spenders for fish and marine products are Region XII, Region III, and BARMM. 
#Their main source of income are other sources of income and entrepreneurial activities. 
sns.scatterplot(y="Region", x="Total Fish and  marine products Expenditure", data=df, hue="Main Source of Income")
```

```python
#Finding 8: The top 3 highest spenders for vegetables are NCR, Region VIII, and BARMM. 
#Their main source of income are wage/salaries and entrepreneurial activities. 
sns.scatterplot(y="Region", x="Vegetables Expenditure", data=df, hue="Main Source of Income")
```

```python
#Finding 9: The top 3 highest spenders for fruits are Region VII, NCR, and BARMM. 
#Their main source of income are wage/salaries and entrepreneurial activities. 
sns.scatterplot(y="Region", x="Fruit Expenditure", data=df, hue="Main Source of Income")
```
